```python
from typing import Dict, List
from collections import defaultdict

lst = ["add", "delete", "edit", "a.add", "a.delete", "c.add", "a.edit", "a.b.add", "a.b.edit", "a.b.d.edit"]
# lst = [_acs.split('.') for _acs in lst]

def gen_dict_from_acl(lst: List[str]) -> dict:
    _dict = defaultdict(list)

    for acs in lst:
        *prefix, verb = acs.split(".")
        if not prefix:
            _dict["."].append(verb)
        else:
            obj = _dict
            for bit in prefix:
                obj = obj.setdefault(bit, defaultdict(list))
            obj["."].append(verb)



    def default_to_regular(_dict: defaultdict):
        if isinstance(_dict, defaultdict):
            _dict = {k: default_to_regular(v) for k, v in _dict.items()}
        return _dict
    
    return default_to_regular(_dict)



cluster = gen_dict_from_acl(lst)
```

```python
from typing import Iterable
import functools

def evaluate_cluster(cluster: dict, title: str):
    index = 0

    def evaluate_cluster_rec(cluster: dict, title: str):
        nonlocal index
        cluster_values = {}
        for key, value in cluster.items():
            if key == ".":
                if isinstance(value, Iterable):
                    for val in value:
                        cluster_values[f"{title}{key}{val}"] = 1 << index
                        index = index + 1
                else:
                    raise Exception("root level should have list of str as acl")
            else:
                subtitle = f"{title}.{key}"
                fills = evaluate_cluster_rec(cluster[key], title=subtitle)
                cluster_values = {**cluster_values, **fills}

        return cluster_values

    return evaluate_cluster_rec(cluster, title)
```

```python
lst = ["add", "delete", "edit", "a.add", "a.delete", "c.add", "a.edit", "a.b.add", "a.b.edit", "a.b.d.edit"]
cluster = gen_dict_from_acl(lst)
evaluate_cluster(cluster, title="resource")
```
