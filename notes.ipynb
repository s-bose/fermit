```python
def find_subresource_string(string: str) -> Optional[str]:
    match = re.match(r"\w+\.\w+(\.\w+)+", string)
    if match:
        return match.group(0)
    return None

def split_resource_string(string: str) -> List[str]:
    return string.split(".")

s = find_subresource_string("product.description.add")
split_resource_string(s)
```

```python
from typing import Dict, List
from collections import defaultdict

lst = ["add", "delete", "edit", "a.add", "a.delete", "c.add", "a.edit", "a.b.add", "a.b.edit", "a.b.d.edit"]
# lst = [_acs.split('.') for _acs in lst]

def gen_dict_from_acl(lst: List[str]) -> dict:
    d = defaultdict(list)

    for acs in lst:
        *prefix, verb = acs.split(".")
        if not prefix:
            d["."].append(verb)
        else:
            obj = d
            for bit in prefix:
                obj = obj.setdefault(bit, defaultdict(list))
            obj["."].append(verb)



    def default_to_regular(_dict: defaultdict):
        if isinstance(_dict, defaultdict):
            _dict = {k: default_to_regular(v) for k, v in d.items()}
        return _dict
    
    return default_to_regular(d)



gen_dict_from_acl(lst)
```
